## Preprocessing Data

```python
# Define which columns are categorical and continuous
categorical_cols = ['Course', 'Gender', 'Stress_Level', 'Depression_Score', 'Anxiety_Score','Sleep_Quality', 
               'Physical_Activity', 'Diet_Quality', 'Social_Support','Relationship_Status', 'Substance_Use', 
               'Counseling_Service_Use', 'Family_History', 'Chronic_Illness','Extracurricular_Involvement', 
               'Residence_Type']
continuous_cols = ['Age', 'Semester_Credit_Load']

# Create a column transformer for preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), continuous_cols),  # Scale continuous variables
        ('cat', OneHotEncoder(), categorical_cols)    # One-hot encode categorical variables
    ])

# Create a pipeline that applies the preprocessing steps
preprocess_pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Transform the data
transformed_data = preprocess_pipeline.fit_transform(df)
```

```python
# You can obtain the column names for the one-hot encoded categorical variables from the one-hot encoder
categorical_encoder = preprocess_pipeline.named_steps['preprocessor'].named_transformers_['cat']
one_hot_feature_names = categorical_encoder.get_feature_names(input_features=categorical_cols)

# Combine the one-hot encoded features with the scaled continuous variables
all_feature_names = list(continuous_cols) + list(one_hot_feature_names)

# Create a DataFrame using the transformed data and feature names
transformed_df = pd.DataFrame(transformed_data, columns=all_feature_names)
```

```python
transformed_df['Age']
```

## Train_Test_Split

```python
X = transformed_df
y = df['CGPA']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=333)
```

## Dummy Model Regressor

```python
# Instantiate dummy model that will always predict majority class
dummy_model = DummyRegressor(strategy="mean")
dummy_model.fit(X_train, y_train)
y_pred = dummy_model.predict(X_train)

mean_squared_error(y_train, y_pred)
```

```python
scores = cross_val_score(dummy_model, X_train, y_train, scoring='neg_mean_squared_error')

# Calculate the mean and standard deviation of the scores
mean_score = -scores.mean()
std_score = scores.std()

print(f'Mean Squared Error (MSE) across {5}-fold CV: {mean_score:.2f}')
print(f'Standard Deviation of MSE: {std_score:.2f}')
```
